```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
# Gerando uma massa de dados:
x, y = make_regression(n_samples=200, n_features=1, noise=30) # quant_dados, quant_variaveis, ruido
# Mostrando o gráfico
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
# Criação do modelo
modelo = LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
modelo.intercept_ #coef linear b -> y=mx+b
```

```python
modelo.coef_ #coef angular m -> y=mx+b
```

```python
# Mostrando o resultado
import numpy as np
plt.scatter(x,y)
xreg = np.arange(-3,3,1)
plt.plot(xreg, 75.09490594*xreg-1.167, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste,y_treino, y_teste = train_test_split(x, y, test_size=0.30)
```

```python
modelo.fit(x_treino, y_treino)
```

```python
resultado = modelo.score(x_teste, y_teste)
```

```python
print(resultado)
```

```python
modelo.intercept_ #coef linear b -> y=mx+b
```

```python
modelo.coef_ #coef angular m -> y=mx+b
```
